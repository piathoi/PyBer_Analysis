```python
#import dependencies
import matplotlib.pyplot as plt
```

```python
#set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot - line chart
plt.plot(x_axis, y_axis)
```

```python
#switch from using MATLAB to the object_oriented interface method, use ax.plot() instead of plt.plot()
#to create figure: use fig, ax = plt.subplots()
#to define axes, use ax.plot()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
#cell 6 can be written like this:
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
#create the plot
plt.plot(x_axis, y_axis)

#plt.show() looks for all the active figure objects and opens a window that displays them is you have 2 or more sets of data to plot
plt.show()
```

```python
#create the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker = "*", color = "blue", linewidth = 2, label='Boston')

#create labels for the x and y axes
plt.xlabel("date")
plt.ylabel("fare($)")

#set the y limit between 0 and 45
plt.ylim(0,45)

#create a title
plt.title("PyBer Fare by Month")

#add a grid
plt.grid()

#add a legend
plt.legend()

#use slim() to show months Jan thro June on graph
plt.xlim("Jan","June") 
# plt.xlim(0,5) is okay too

#change the figure size to 20 pixels wide by 1- pixels high using figure
plt.figure(figsize=(20,10)) 
# plt.subplots(figsize=(20,10)) is okay too
```

# Scatter Plot

## Scatter Plot using MATLAB method

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#using plt.plot() doesnt work because it will create a line plot.
#insert lowercase "o" 
plt.plot(x_axis, y_axis, 'o')
```

```python
#create scatter plot using plt.scatter can also works
plt.scatter(x_axis, y_axis)
```

```python
#create the plot 
plt.scatter(y_axis, x_axis, color = "red", label='Chicago')

#create labels for the x and y axes
plt.ylabel("Fare($)")
plt.xlabel("Date")

#create a title
plt.title("PyBer Fare by Month")

#add the legend
plt.legend()
```
